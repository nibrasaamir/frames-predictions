```python
from google.colab import drive
import os
import pandas as pd


train_folder_path = '/content/drive/MyDrive/train'
train_csv_path = '/content/drive/MyDrive/train.csv'


if os.path.exists(train_folder_path):
    print(f"Train folder found at '{train_folder_path}'.")
else:
    print(f"Train folder not found at '{train_folder_path}'. Please check the path.")


try:
    df_train = pd.read_csv(train_csv_path)
    print("Successfully loaded train.csv")
    print(df_train.head())
except FileNotFoundError:
    print(f"Error: File not found at {train_csv_path}. Please check the path.")
except pd.errors.EmptyDataError:
    print(f"Error: File at {train_csv_path} is empty.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

import os


target_labels = ['BrushingTeeth', 'Biking', 'PushUps', 'PullUps', 'BenchPress', 'JumpingJack']

train_filtered_df = df_train[df_train['label'].isin(target_labels)]
print("Filtered labels:", train_filtered_df['label'].unique())

train_base_path = '/content/drive/MyDrive/train'

train_filtered_df['clip_path'] = train_filtered_df.apply(
    lambda row: os.path.join(train_base_path, row['label'], row['clip_name'] + '.avi'), axis=1
)

print("\nVerifying clip paths:")
for idx, row in train_filtered_df.iterrows():
    exists = os.path.exists(row['clip_path'])
    print(f"{row['clip_path']} - Exists: {exists}")


train_df = train_filtered_df.copy()
#############################################################################################################################################

import tensorflow as tf
from tensorflow.keras import layers, models

# Check GPU availability
device = '/GPU:0' if tf.config.list_physical_devices('GPU') else '/CPU:0'
print(f"Using device: {device}")


FRAME_HEIGHT, FRAME_WIDTH = 64, 64
NUM_INPUT_FRAMES = 10
NUM_PREDICT_FRAMES = 5
BATCH_SIZE = 4
EPOCHS = 30
LEARNING_RATE = 0.001  

# Custom SSIM Loss
def ssim_loss(y_true, y_pred):
    return 1 - tf.reduce_mean(tf.image.ssim(y_true, y_pred, max_val=1.0))


def combined_loss(y_true, y_pred):
    return ssim_loss(y_true, y_pred) + tf.reduce_mean(tf.keras.losses.MAE(y_true, y_pred))


def ssim_metric(y_true, y_pred):
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(y_pred, tf.float32)
    return tf.reduce_mean(tf.image.ssim(y_true, y_pred, max_val=1.0))


def create_conv_lstm_model(input_shape, num_predict_frames):
    inputs = layers.Input(shape=input_shape)

    #encoder
    encoder = layers.ConvLSTM2D(64, (3, 3), padding='same', return_sequences=True)(inputs)
    encoder = layers.BatchNormalization()(encoder)
    encoder = layers.ConvLSTM2D(64, (3, 3), padding='same', return_sequences=True)(encoder)
    encoder = layers.BatchNormalization()(encoder)

    #decoder
    decoder = layers.ConvLSTM2D(64, (3, 3), padding='same', return_sequences=True)(encoder)
    decoder = layers.BatchNormalization()(decoder)
    decoder = layers.ConvLSTM2D(64, (3, 3), padding='same', return_sequences=True)(decoder)
    decoder = layers.BatchNormalization()(decoder)

   
    output = layers.Conv3D(3, (3, 3, 3), activation='sigmoid', padding='same')(decoder)

    
    output = layers.Lambda(lambda x: x[:, -num_predict_frames:, :, :, :])(output)

    model = models.Model(inputs=inputs, outputs=output)
    return model


input_shape = (NUM_INPUT_FRAMES, FRAME_HEIGHT, FRAME_WIDTH, 3)
model = create_conv_lstm_model(input_shape, NUM_PREDICT_FRAMES)


model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE),
    loss=combined_loss,
    metrics=['mae', ssim_metric]  
)


model.summary()


from tensorflow.keras.utils import Sequence
import cv2
import numpy as np

class VideoDataLoader(Sequence):
    def __init__(self, dataframe, batch_size):
        self.dataframe = dataframe.reset_index(drop=True)
        self.batch_size = batch_size
        self.indices = np.arange(len(self.dataframe))

    def __len__(self):
        return int(np.ceil(len(self.dataframe) / self.batch_size))

    def __getitem__(self, idx):
        batch_indices = self.indices[idx * self.batch_size: (idx + 1) * self.batch_size]
        batch_clips = self.dataframe.iloc[batch_indices]
        input_frames, output_frames = [], []

        for _, row in batch_clips.iterrows():
            video_path = row['clip_path']
            cap = cv2.VideoCapture(video_path)
            frames = []

            
            while True:
                ret, frame = cap.read()
                if not ret:
                    break
                frame = cv2.resize(frame, (FRAME_WIDTH, FRAME_HEIGHT))
                frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB) 
                frame = frame / 255.0  
                frame = frame.astype(np.float32)  
                frames.append(frame)
            cap.release()

          
            if len(frames) >= NUM_INPUT_FRAMES + NUM_PREDICT_FRAMES:
                
                input_frame_indices = []
                remaining_frames = len(frames) - NUM_PREDICT_FRAMES
                idx_frame = 0

                while len(input_frame_indices) < NUM_INPUT_FRAMES and idx_frame < remaining_frames:
                    input_frame_indices.append(idx_frame)
                    if len(input_frame_indices) % 2 == 1:
                        idx_frame += 6  # my skip 5 logic
                    else:
                        idx_frame += 1  

                   
                    if idx_frame >= remaining_frames:
                        input_frame_indices = list(range(remaining_frames - NUM_INPUT_FRAMES, remaining_frames))
                        break

              
                input_frames.append([frames[i] for i in input_frame_indices])
                output_frames.append(frames[NUM_INPUT_FRAMES:NUM_INPUT_FRAMES + NUM_PREDICT_FRAMES])

        return np.array(input_frames, dtype=np.float32), np.array(output_frames, dtype=np.float32)


train_loader = VideoDataLoader(train_df, BATCH_SIZE)
# trainingg
with tf.device(device):
    history = model.fit(train_loader, epochs=EPOCHS)

import tensorflow as tf
from tensorflow import keras

keras.config.enable_unsafe_deserialization()

from tensorflow.keras.models import load_model

model.save("conv_lstm_model.keras")
print("Model saved successfully as 'conv_lstm_model.keras'.")

loaded_model = load_model("conv_lstm_model.keras", custom_objects={
    'combined_loss': combined_loss,
    'ssim_loss': ssim_loss,
    'ssim_metric': ssim_metric 
})
print("Model loaded successfully.")

loaded_model.summary()

import matplotlib.pyplot as plt

test_data = train_df.head(5)

input_frames = []

for _, row in test_data.iterrows():
    video_path = row['clip_path']
    cap = cv2.VideoCapture(video_path)
    frames = []

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame = cv2.resize(frame, (FRAME_WIDTH, FRAME_HEIGHT))
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB) 
        frame = frame / 255.0 
        frame = frame.astype(np.float32) 
        frames.append(frame)
    cap.release()

    input_frame_indices = []
    remaining_frames = len(frames) - NUM_PREDICT_FRAMES
    idx_frame = 0

    while len(input_frame_indices) < NUM_INPUT_FRAMES and idx_frame < remaining_frames:
        input_frame_indices.append(idx_frame)
        if len(input_frame_indices) % 2 == 1:
            idx_frame += 6  
        else:
            idx_frame += 1 

        if idx_frame >= remaining_frames:
            input_frame_indices = list(range(remaining_frames - NUM_INPUT_FRAMES, remaining_frames))
            break

    input_frames.append([frames[i] for i in input_frame_indices])

input_frames = np.array(input_frames, dtype=np.float32)

predicted_frames = loaded_model.predict(input_frames)

def display_frames(input_frames, predicted_frames):
    num_samples = input_frames.shape[0]
    for i in range(num_samples):
        plt.figure(figsize=(20, 4))

        for j in range(NUM_INPUT_FRAMES):
            plt.subplot(2, NUM_INPUT_FRAMES, j + 1)
            plt.imshow(input_frames[i][j])
            plt.axis('off')
            if j == 0:
                plt.title("Input Frames", fontsize=16)

        for j in range(NUM_PREDICT_FRAMES):
            plt.subplot(2, NUM_INPUT_FRAMES, NUM_INPUT_FRAMES + j + 1)
            plt.imshow(predicted_frames[i][j])
            plt.axis('off')
            if j == 0:
                plt.title("Predicted Frames", fontsize=16)

        plt.show()


display_frames(input_frames, predicted_frames)

import os
import cv2

output_dir = '/content/drive/MyDrive/train1_answers_conv_lstm'
os.makedirs(output_dir, exist_ok=True)

test_data_save = train_df.tail(5)
input_frames_save = []

for _, row in test_data_save.iterrows():
    video_path = row['clip_path']
    cap = cv2.VideoCapture(video_path)
    frames = []

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame = cv2.resize(frame, (FRAME_WIDTH, FRAME_HEIGHT))
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB) 
        frame = frame / 255.0  
        frame = frame.astype(np.float32)  
        frames.append(frame)
    cap.release()

    input_frame_indices = []
    remaining_frames = len(frames) - NUM_PREDICT_FRAMES
    idx_frame = 0

    while len(input_frame_indices) < NUM_INPUT_FRAMES and idx_frame < remaining_frames:
        input_frame_indices.append(idx_frame)
        if len(input_frame_indices) % 2 == 1:
            idx_frame += 6  # logic to skip 5
        else:
            idx_frame += 1  

        if idx_frame >= remaining_frames:
            input_frame_indices = list(range(remaining_frames - NUM_INPUT_FRAMES, remaining_frames))
            break

    input_frames_save.append([frames[i] for i in input_frame_indices])

input_frames_save = np.array(input_frames_save, dtype=np.float32)


predicted_frames_save = loaded_model.predict(input_frames_save)


def save_frames(input_frames, predicted_frames, index, output_dir):
  
    for j in range(NUM_INPUT_FRAMES):
        input_frame = input_frames[index][j] * 255.0  
        input_frame = input_frame.astype(np.uint8)
        input_filename = os.path.join(output_dir, f"input_{index+1}_frame_{j+1}.png")
        cv2.imwrite(input_filename, cv2.cvtColor(input_frame, cv2.COLOR_RGB2BGR))

 
    for j in range(NUM_PREDICT_FRAMES):
        predicted_frame = predicted_frames[index][j] * 255.0  
        predicted_frame = predicted_frame.astype(np.uint8)
        predicted_filename = os.path.join(output_dir, f"predicted_{index+1}_frame_{j+1}.png")
        cv2.imwrite(predicted_filename, cv2.cvtColor(predicted_frame, cv2.COLOR_RGB2BGR))


for i in range(input_frames_save.shape[0]):
    save_frames(input_frames_save, predicted_frames_save, i, output_dir)

print(f"Frames saved successfully in '{output_dir}' folder.")
```
